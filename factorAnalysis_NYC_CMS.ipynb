#Scree test to find eigenvalues


```python
#EXPLORATORY FACTOR ANALYSIS with 10 randomly guessed 10 factors
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(10, rotation=None)
fa.fit(df3)
```

```python
#GET EIGENVALUES
ev,v = fa.get_eigenvalues()
print(ev) #it is the original_eigen_values
print(v) # it is the common_factor_eigen_values
```

```python
# SCREEPLOT to compare eigenvalues
plt.scatter(range(1,df3.shape[1]+1),ev)
plt.plot(range(1,df3.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

#Based on the scree plot, I will choose a solution with 7 or 8 factors. Lets test them out

#Lets try 8 Factors




###Loadings without rotation





```python
# Create factor analysis object and perform factor analysis
fa8 = FactorAnalyzer(8, rotation=None)
fa8.fit(df3)
```

```python
fa8.loadings_
```

```python
#Create Factot Loadings/ Component Matrix
loadings8 = pd.DataFrame(fa8.loadings_, columns=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4', 'Factor 5', 'Factor 6', 'Factor 7', 'Factor 8'], index=df3.columns)
print('Factor Loadings/ Component Matrix \n%s' %loadings8)
```

```python
#create heatmap of factor loadings
sns.heatmap(loadings8, vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),annot=True)
plt.title('Factor Loadings for 8 factors without rotation')
plt.show()
```

```python
fa8.get_communalities()
```

```python
#compare communalities
communalities = pd.DataFrame(fa8.get_communalities(), columns=['Extraction'], index=df3.columns)
print('Communalities \n%s' %communalities)
```

But most of the variables are clustered in factor 1. So to balance this out, we use rotational matrix using Varimax rotation

###Loading with varimax rotation using minres method

---



```python
# Create factor analysis object and perform factor analysis
fa_rot = FactorAnalyzer(8, rotation="varimax", method='minres', use_smc=True)
fa_rot.fit(df3)
```

```python
#Create Factot Loadings/ Component Matrix
loadings_rot = pd.DataFrame(fa_rot.loadings_, columns=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4', 'Factor 5', 'Factor 6', 'Factor 7', 'Factor 8'], index=df3.columns)
print('Factor Loadings/ Component Matrix \n%s' %loadings_rot)
```

```python
#create heatmap of factor loadings
sns.heatmap(loadings_rot, vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),annot=True)
plt.title('Factor Loadings with 8 factors using varimax and minres method')
plt.show()
```

```python
# Get variance of each factors
fa_rot.get_factor_variance()
```

```python
fa_rot.get_communalities()
#compare communalities
communalities_rot= pd.DataFrame(fa_rot.get_communalities(), columns=['Extraction'], index=df3.columns)
print('Communalities \n%s' %communalities_rot)
```
